```python



data['Duration_Seconds'] = data['Duration_MS'] / 1000

# Izračunavanje minuta i sekundi
data['Minutes'] = data['Duration_Seconds'] // 60
data['Seconds'] = data['Duration_Seconds'] % 60

# Kombinovanje minuta i sekundi u format 'M.SS'
data['Duration_Minutes'] = data['Minutes'] + (data['Seconds'] / 100)
duration_minutes = data['Duration_Minutes']
duration_minutes
```

```python
#Trajanje pjesama prosjecno

def correct_minutes(duration_minutes):
    # Zaokruživanje na dve decimale
    corrected_minutes = round(duration_minutes + 0.01, 2)
    # Ako decimalni deo prelazi 0.59, povećajte celobrojnu vrednost za 1
    if corrected_minutes - int(corrected_minutes) >= 0.6:
        corrected_minutes = int(corrected_minutes) + 1
    return corrected_minutes

duration_minutes= data['Duration_MS']/60000

sns.set_theme()
plt.figure(figsize=(12,8))
sns.histplot(duration_minutes, bins= 30, kde= True, edgecolor= 'white', linewidth= 1.2, alpha =0.6)

plt.title('Trajanje pjesama u minutama', fontsize=16)
plt.xlabel('Duzina', fontsize=14)
plt.ylabel('Gustina', fontsize=14)

mean_duration = correct_minutes(duration_minutes.mean())
plt.axvline(x=mean_duration, color='red', linestyle='--', label=f'Prosjecno trajanje ({mean_duration:.2f})')

plt.legend()
plt.show()
```

```python
# Prosek trajanja pesama po godinama
average_duration_per_year = data.groupby('Year')['Duration_Seconds'].mean()

# Konvertovanje proseka trajanja iz sekundi u minute
average_duration_per_year_minutes = average_duration_per_year / 60

# Grafikon prosečnog trajanja pesama po godinama
plt.figure(figsize=(12, 8))
plt.plot(average_duration_per_year.index, average_duration_per_year_minutes.values, marker='o', linestyle='-', color='blue', linewidth=2)
plt.title('Prosjecno trajanje pjesama po godinama', fontsize=16)
plt.xlabel('Godina', fontsize=14)
plt.ylabel('Prosjecno trajanje (sek)', fontsize=14)
plt.xticks(average_duration_per_year.index.astype(int))
plt.grid(True)

y_ticks = [round(value, 2) for value in average_duration_per_year_minutes.values if round(value, 2) % 0.5 == 0]
plt.yticks(y_ticks)

plt.show()
```

```python
#Matrica korelacije

numeric_columns = ['Duration_MS', 'Popularity', 'Danceability', 'Energy', 'Key', 'Loudness', 
                   'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo']

correlation_matrix = data[numeric_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='YlOrRd', fmt=".2f", linewidths=.5)
plt.title('Matrica korelacije', fontsize=16)
plt.show()
```
